### Dealing With the missing Values 

```python
# Creating a nan value Dataframe
sparse = pd.DataFrame(
    {
    'a': [1,2,3],
    'b': [np.nan,4,np.nan],
    'c': [5, np.nan,np.nan]}

)
```

```python
sparse
```

```python
# checking nan value in the sparse matrix
sparse.isna()
```

```python
# fill nan value with 0
sparse.fillna(0)
```

```python
# dropping na colun having nan value in it 
sparse.dropna(axis = 1)
```

### Group By

### Merging and Joining

```python
contacts = pd.DataFrame(
    {
     'name' :['Muzzi','Amna'] ,
     'Tel'  :['+91 999 999 99 99', '+91 888 888 88 88'] 
    }
)
```

```python
contacts
```

```python
exam_df.merge(contacts ,left_on = 'name', right_on = 'name')# default inner join
```

```python
exam_df.merge(contacts ,left_on = 'name', right_on = 'name', how = 'outer')
```
