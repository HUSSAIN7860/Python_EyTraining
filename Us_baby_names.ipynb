### Introduction:
We are going to use a subset of US Baby Names from Kaggle.
In the file it will be names from 2004 until 2014

```python
# import the necessary library

import numpy as np
import pandas as pd
```

```python
# uploading dataset to this notebook
url = "https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv"

df = pd.read_csv(url)

df.info()
```

```python
# assign df to baby_names
baby_names = df.copy()
```

```python
# first 10 rows of the dataset
baby_names.head(10)
```

```python
# deleting the coloumn 'Unnamed :0' and 'Id'

baby_names = baby_names.drop(['Unnamed: 0','Id'],axis = 1)

baby_names.head()
```

```python
# Is there more male or female names in the dataset?
baby_names['Gender'].value_counts()
```

```python
# deleting the year colonm

del baby_names['Year']


# Group the dataset by name and assign to names



names = baby_names.groupby('Name').sum()

# sort the value in descending order
names.sort_values('Count',ascending = 0).head()
```

```python
# different names in the data set

len(names)
```

```python
# name with most occurances
#names.Count.idxmax()

# or name with Count
names[names.Count == names.Count.max()]
```

```python
# list occurances
len(names[names.Count == names.Count.min()])
```

```python
# calulate median of occurancess
names[names.Count == names.Count.median()]
```

```python
# standard deviation 
names.Count.std()
```

```python
# summary 
names.describe()
```
