### Import Library Here :

```python
# import Libraries here 

# pandas - a powerful data analysis and manipulation library for Python
import pandas as pd 
#Seaborn is a Python data visualization library based on matplotlib
import seaborn as sns


# print the graph in the notebook itself
%matplotlib inline 

# set seaborn style to white
sns.set_style("white")
```

### Import Dataset to this notebook by using this url :

```python
# url 
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/07_Visualization/Tips/tips.csv'

# read comma seprated values here with the help of pandas
df = pd.read_csv(url)

# print some colunm 
df.head()
```

### Here we have to remove column Unnamed :0 from data frame

```python
del df['Unnamed: 0']

df.head(10)
```

### Ploting Histogram from tip to total_bill

```python
# create histogram 
tip_totalbill = sns.distplot(df.total_bill)

# rename the axis
tip_totalbill .set(xlabel = ' Value' , ylabel = 'Frequency' , title = 'total bill' )

# remove the border 
sns.despine()
```

### Creating Scatter plot to check the relationship of total bill and tips 

```python
# for this we create joint plot here 

sns.jointplot(x = "total_bill" , y = "tip" , data = df)
```

### Create one image with the relationship of total_bill, tip and size.

```python
# for this we create pairplot here 

sns.pairplot(df)
```

### showing the reltaionship between days and total_bill values 

```python
# for this we create stripplot here :

sns.stripplot( x ='day' , y ='total_bill',data = df ,jitter = True  )
# using jiiter for single point data set 
```

###Create a scatter plot with the day as the y-axis and tip as the x-axis, differ the dots by sex

```python
# fro this stripplot
sns.stripplot(x = 'tip' ,y ='day' , hue = 'sex' ,data = df , jitter = True )
# hue is used for differentiationg the datapoint 
```

### Create a box plot presenting the total_bill per day differetiation the time (Dinner or Lunch)

```python
# for this creating the boxplot
sns.boxplot(x = 'day', y = 'total_bill' , data = df , hue = 'time')
```

###  Create two histograms of the tip value based for Dinner and Lunch. They must be side by side.

```python
# import matplotlib 

import matplotlib.pyplot as plt

# better seaborn style
sns.set(style = "ticks")

# create facegrid
m = sns.FacetGrid(df, col = 'time')
m.map(plt.hist , "tip");
```

### Create two scatterplots graphs, one for Male and another for Female, presenting the total_bill value and tip relationship, differing by smoker or no smoker
They must be side by side.

```python
m = sns.FacetGrid(df,col = 'sex' ,hue = 'smoker')
m.map(plt.scatter , 'total_bill' , 'tip' , alpha = 0.8)


# this show the color of smoker
m.add_legend();
```
